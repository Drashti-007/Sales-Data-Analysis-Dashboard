```python
df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
df.info()
```

```python
# Basic Business Matrics

# Total Sales
total_sales = df["Sales"].sum()
print(total_sales)
```

```python
# Total Profit
total_profit = df["Profit"].sum()
print(total_profit)
```

```python
# Profit Margin
profit_margin =  (total_profit / total_sales) * 100
print(profit_margin)
```

```python
# Average Sales
average_sales = df["Sales"].mean() 
print(average_sales)
```

```python
# Highest Sales
highest_sales = df["Sales"].max()
print(highest_sales)
```

```python
# Region Sales
region_sales = (
    df.groupby("Region")["Sales"]
    .sum()
    .sort_values(ascending=False)
)
print(region_sales)
```

```python
# Region Profit
region_profit = df.groupby("Region")["Profit"].sum()
print(region_profit)
```

```python
# Category Sales
category_sales = df.groupby("Category")["Sales"].sum()
print(category_sales)
```

```python
# Category Profit
category_profit = df.groupby("Category")["Profit"].sum()
print(category_profit)
```

```python
# sorting
category_sales = (
    df.groupby("Category")["Sales"]
    .sum()
    .sort_values(ascending = False)
)
print(category_sales)
```

```python
# top products by sales
top_products = (
    df.groupby("Product Name")["Sales"]
    .sum()
    .sort_values(ascending = False)
)
top_products.head(10)
```

```python
top_products.tail(10)
```

```python
# timesseries analysis
# Extract year and month from order date

df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month_name()

df[["Order Date", "Year", "Month"]].head()
```

```python
yearly_sales = (
    df.groupby("Year")["Sales"]
    .sum()
)
yearly_sales
```

```python
monthly_sales = (
    df.groupby("Month")["Sales"]
    .sum()
)
monthly_sales
```

```python
month_order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
```

```python
# converting month column to ordered categorical 
df["Month"] = pd.Categorical(
    df["Month"],
    categories=month_order,
    ordered=True
)
```

```python
monthly_sales = (
    df.groupby("Month")["Sales"]
    .sum()
    .sort_index()
)
monthly_sales
```

```python
# Visualization
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    monthly_sales.index,
    monthly_sales.values,
    marker='o', # o=circle, s=square, ^ = triangle, *=star, x=cross, +=plus
    markersize=8, 
    linewidth=3
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    category_sales.index,
    category_sales.values,
)

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    region_sales.index,
    region_sales.values
)

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()
```

```python
plt.figure(figsize=(7,7))

plt.pie(
    category_sales.values,
    labels=category_sales.index,
    autopct="%1.1f%%"
)

plt.title("Sales Distribution by Category")

plt.show()
```
